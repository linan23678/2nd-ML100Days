# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
## Relu 數學式

def Relu(x):
    return abs(x)*(x>0)

def dRelu(x):
    return ( 1 * ( x > 0) )


## x範圍
x = np.linspace(-3,3,100)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, Relu(x), 'b') ## , label='linspace(-10,10,10)'
plt.plot(x, dRelu(x), 'r') ##, label='linspace(-10,10,10)'

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('Relu Function &　dRelu')

# 顯現 the Sigmoid formula
plt.text(-0.5, 2.5, r'$f(x) = abs(x)*(x>0)$', fontsize=15)
plt.text(-2.4, 0.5, r'$f(x) = (1*(x>0))$', fontsize=15)

# #resize the X and Y axes
# plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
# plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 
# create the graph
plt.show()
```

```python
from numpy import exp as e
## tanh 數學式
def tanh(x):
    return (e(x)-e(-x))/(e(x)+e(-x))


## x範圍
x = np.linspace(-10,10,100)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, tanh(x), 'b') ## , label='linspace(-10,10,10)'

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('Tanh Function')

# 顯現 the Sigmoid formula
plt.text(3, 0.5, r'$f(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$', fontsize=20)

# #resize the X and Y axes
# plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
# plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 
# create the graph
plt.show()
```
