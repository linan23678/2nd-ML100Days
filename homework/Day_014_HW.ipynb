# 使用 application_train.csv, 根據不同的 HOUSETYPE_MODE 對 AMT_CREDIT 繪製 Histogram     ¶

```python
app_train_notull_HOUSE =   app_train[~app_train['HOUSETYPE_MODE'].isnull()]['HOUSETYPE_MODE'] ## 有數值

app_train_House = app_train_notull_HOUSE.unique()
```

```python
unique_house_type = app_train_House    

for i in unique_house_type:
    print(i)
    
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = app_train_House
nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    

    """
    Your Code Here
    """
    app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i],'AMT_CREDIT'].hist( color = 'b') ## array使用[]

    plt.title(str(unique_house_type[i]))
plt.show()    
```
